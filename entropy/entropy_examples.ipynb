# Entropy of a Bernoulli RV as a function of $\theta$

```python
import matplotlib.pyplot as plt
import numpy as np
theta = np.linspace(1e-16,1-1e-16,101)
H_bernoulli = -(theta*np.log2(theta) + (1-theta)*np.log2(1-theta))
plt.plot(theta,H_bernoulli)
```

# Entropy of a Binomial RV as a function of $\theta$ and $n$

```python
from scipy.stats import binom
n = 2
theta = np.linspace(1e-16,1-1e-16,101)
x = np.linspace(0,n,n+1)
H = []

for t in theta:
    p = binom.pmf(x,n,t)
    H.append(-(p*np.log2(p)).sum())
H = np.array(H)

plt.plot(theta, H)
```

# Entropy of Bernoulli trials as a function of $\theta$ and $n$

```python
def b_trial(theta,X):
    n = len(X)
    return theta**sum(X)*(1-theta)**(len(X) - sum(X))

H = 0
for x1 in range(2):
    for x2 in range(2):
        H -= b_trial(theta,[x1,x2])*np.log2(b_trial(theta,[x1,x2]))

plt.plot(theta,H)
```
